```python
# Area of a circle
def area_of_circle(r):
    return 3.14*r*r

radius=float(input('Enter radius : '))
print("Area of circle = ",area_of_circle(radius))
```

```python
# Find the largest among three numbers

def largest(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3
        
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

result = largest(num1, num2, num3)
print("Largest number is:", result)
```

```python
# Length of string without using built-in function.
def find_length(text):
    count = 0

    for char in text:
        count += 1

    return count
char = str(input("Enter your string"))

result = find_length(char)
print(result)
```

```python
# Find Prime Numbers
def is_prime(num):

    if num <= 1:
        return False

    for i in range(2, num):
        if num % i == 0:
            return False

    return True


number = int(input("Enter a number: "))

if is_prime(number):
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
#Recursive Fibonacci Number
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input("Enter n: "))
print("Fibonacci =", fibonacci(n))
```

```python
#Recursive String Reverse
def rev(s):
    if s == "":
        return s
    return rev(s[1:]) + s[0]

s = input("Enter string: ")
print("Reverse =", rev(s))
```

```python
#Sum of Numbers from 1 to n Using for Loop
n = int(input("Enter n: "))

s = 0

for i in range(1, n + 1):
    s += i

print("Sum =", s)
```

```python
##Count Even Numbers in a List
numbers = [2, 5, 8, 11, 14, 20]

count = 0

for num in numbers:
    if num % 2 == 0:
        count += 1

print("Even numbers =", count)
```

```python
# program to find of number 1 to n using for loop 
n = int(input("Enter a number: "))

total = 0

for i in range(1, n + 1):
    total = total + i

print("Sum =", total)
```
